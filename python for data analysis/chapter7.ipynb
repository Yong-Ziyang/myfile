```python
import pandas as pd
import numpy as np
from pandas import DataFrame,Series
```

```python
# merge函数使用
data1=DataFrame({'key':['b','a','c','c','b','a'],
               'value1':range(6)})
data2=DataFrame({'key':['a','b','d'],
                'value2':range(3)})
```

```python
pd.merge(data1,data2)# 这个一种多对一的合并,双方都有的key才会合并,这个python是将双方都有的key作为合并项
```

```python
pd.merge(data1,data2,on='key') # 最好显式的指定一下是用那一列作为合并列
```

```python
# key不一样怎么做
data3=DataFrame({'key1':['b','a','c','e','b','a'],
               'value1':range(6)})
data4=DataFrame({'key2':['a','b','a','b','c','d'],
                'value2':range(6)})
```

```python
pd.merge(data3,data4,right_on='key2',left_on='key1') 
# python 默认是inner连接，其他方式还有'left','right','outer'链接方式
```

```python
pd.merge(data3,data4,right_on='key2',left_on='key1',how='right') # 右链接
```

```python
pd.merge(data3,data4,right_on='key2',left_on='key1',how='left') # 左链接
```

```python
pd.merge(data3,data4,right_on='key2',left_on='key1',how='outer') # 外链接
```

```python
# 处理重复列名
left=DataFrame({'key1':['foo','foo','bar'],
              'key2':['one','two','three'],
              'lval':[1,2,3]})
right=DataFrame({'key1':['foo','foo','bar','bar'],
               'key2':['one','one','one','two'],
               'rval':[4,5,6,7]})
```

```python
pd.merge(left,right,on=['key1','key2'])
```

```python
pd.merge(left,right,on=['key1','key2'],how='outer')
```

```python
# 连接时如何处理重复列名
pd.merge(left,right,on='key1') # 系统会自动将key2加上后缀，也可以自选后缀
```

```python
pd.merge(left,right,on='key1',suffixes=['_left','_right'])
```

```python
pd.merge(left,right,on='key1',suffixes=['_left','_right'],sort=False) # sort 设置成False,在处理大数据时可以提高性能
```
